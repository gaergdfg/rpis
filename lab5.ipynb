```python
# Problem 5a

M = 420
N = 1000
T = 1000

def estimator_1(arr):
    return sum(arr) * 2 / len(arr) - 1

def estimator_2(arr):
    return max(arr) * (len(arr) + 1) / len(arr) - 1

def sample_estimator_1():
    arr = np.random.randint(M, size=N)
    return estimator_1(arr)

def sample_estimator_2():
    arr = np.random.randint(M, size=N)
    return estimator_2(arr)

results_1 = np.array([sample_estimator_1() for i in range(0, T)])
results_2 = np.array([sample_estimator_2() for i in range(0, T)])

average_1 = np.average(results_1)
average_2 = np.average(results_2)

variance_1 = np.var(results_1)
variance_2 = np.var(results_2)

print("1", average_1, variance_1)
print("2", average_2, variance_2)
```

```python
# Problem 2c

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


da = pd.read_csv('us_births_69_88.csv')
#remove error dates
error_dates = [(day, month) for (month,day) in [(2,30), (2,31), (4,31), (6,31), (4,31), (6,31), (9,31), (11,31)]]
da.drop([i for i in da.index if (da.loc[i, 'day'], da.loc[i, 'month']) in error_dates], inplace=True)
births_emp = np.array(da['births'])
births_emp = births_emp * len(births_emp)
births_mean = np.mean(births_emp)


# prepare buckets for sampling
buckets = [(i, -1, births_emp[i]) for i in range(len(births_emp))]

for x in range(len(buckets)):
    for a in range(len(buckets)):
        if births_emp[a] < births_mean:
            for b in range(len(buckets)):
                if births_emp[b] > births_mean:
                    buckets[b] = (buckets[b][0], buckets[b][1], buckets[b][1] - (births_mean - buckets[a][1]))
                    buckets[a] = (buckets[a][0], b, buckets[a][2])
                    break
                    
                    
N = 100000
N = 24 * N
bucket_pos = np.random.randint(0, len(births_emp), N)
bucket_births_threshold = np.random.randint(0, births_mean, N)

days = np.zeros(N)
for i in range(len(bucket_pos)):
    if bucket_births_threshold[i] <= buckets[bucket_pos[i]][2]:
        days[i] = buckets[bucket_pos[i]][0]
    else:
        days[i] = buckets[bucket_pos[i]][1]
        
        
def sample_until_repeat_bucket(days):
    results = []
    currGroup = set()
    for day in days:
        if day in currGroup:
            results.append(len(currGroup))
            currGroup = set()
        else:
            currGroup.add(day)
    return results
```

```python
import scipy.stats

THRESHOLD = 0.05

day_prob = births_emp / len(births_emp)
day_prob = day_prob / np.sum(day_prob)
f_n = N * day_prob

days_obs = []
for i in range(0, 366):
    days_obs.append(np.count_nonzero(days == i))

S = np.sum(((days_obs - f_n) ** 2) / f_n)
print(((days_obs - f_n) ** 2) / f_n)
print(S)

degrees_of_freedom = N - 1
print(scipy.stats.chi2.cdf(S, degrees_of_freedom))

print(scipy.stats.chisquare(days_obs, f_n))
```
