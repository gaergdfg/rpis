```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Part 1
def p_not_0(k):
    return 1 / (4 * abs(k) * (abs(k) + 1))

def p(k):
    return 1 / 2 if k == 0 else p_not_0(k)

x_arr = np.arange(-10, 11)
y_arr = np.vectorize(p)(x_arr)

plt.plot(x_arr, y_arr)

plt.show()

# Part 2 + 3
def cube_rand_sign():
    val = np.random.randint(1, 5)
    
    if val == 1:
        return 1
    elif val == 2:
        return -1
    else: # val == 3 || val == 4
        return 0
    
def sample():
    val = np.random.uniform(0, 1)
    
    return cube_rand_sign() * math.floor(1 / val)

N = 10000

x_arr = np.arange(1, N + 1)
y_arr = np.array([sample() for i in x_arr])
y_arr_sum = np.cumsum(y_arr)
y_arr_avr = y_arr_sum / x_arr

plt.plot(x_arr, y_arr_avr)
plt.show()

y_arr_med = [np.median(y_arr[:i]) for i in x_arr]

plt.plot(x_arr, y_arr_med)
plt.show()
```
